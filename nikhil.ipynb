```python
import cv2
```

 **facial recognition source file of human**

```python
cascPath = "haarcascade_frontalface_default.xml"
faceCascade =  cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
from pathlib import Path
path = '/content/Nikhil_ID.jpeg'
image = cv2.imread(path) 

image_crop = Image.open(path)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

**Face detection code**

```python
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=2,
    minSize=(100, 160),
    flags = cv2.CASCADE_SCALE_IMAGE

)
print("Found {0} faces!".format(len(faces)))
```

**Final Detected face will displayed**

# extra supporting codes for rectangle and co ordinates

```python
gray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
gray = cv.GaussianBlur(gray,(5,5),0)
_, bin = cv.threshold(gray,120,255,1)
bin = cv.dilate(bin, None)
bin = cv.dilate(bin, None)
bin = cv.erode(bin, None)
bin = cv.erode(bin, None)
contours, hierarchy = cv2.findContours(bin, cv.RETR_LIST, cv.CHAIN_APPROX_SIMPLE)

rc = cv.minAreaRect(contours[0])
box = cv.boxPoints(rc)
for p in box:
  pt= (p[0],p[1])
  print(pt)
  cv.circle(im,pt,5,(0,255,0),2)

display(im)
```

```python
for i in box:
    cv2.circle(image,(i[0],i[1]),3, (255,0,0), -1)
    imgplot = plt.imshow(image)
    plt.show()
```

```python
for (x, y, w, h) in faces:

    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)


im_crop = image_crop.crop((x, y, (x+w), (y+h)))

plt.imshow(im_crop)


```
